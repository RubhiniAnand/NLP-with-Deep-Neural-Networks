# What is embedding matrix?
<font size="3">The words separated by the tokenizer method is replaced by their vector values which we had calculated using the pre-trained Word2vec model.<br>In deep learning models (such as LSTM, GRU, or Transformer-based models), an embedding matrix is fed to an embedding layer as part of the model. This allows the model to work with word embeddings directly, which are continuous and more meaningful for machine learning models compared to one-hot vectors or raw text data.<br>An empty marix is created with vocab_size and Vector size of the model(e.g. 6979,100)</font>

# What the embedding marix holds? 
<font size="3"> 1. The loop "for word, i in english_tokenizer.word_index.items():"goes through each word in the english_tokenizer's word index and tries to find the corresponding pre-trained word embedding.<br> 2. If it finds it, the embedding is placed in the matrix at the corresponding index. Otherwise, it moves on without updating the matrix (leaving it filled with zeros for that word).</font>

```python
embedding_vector
```

```python
# Function to create embedding matrix
def create_embedding_matrix(word2vec_model, tokenizer, vocab_size):
    embedding_matrix = np.zeros((vocab_size, word2vec_model.vector_size))
    for word, i in tokenizer.word_index.items():
        try:
            embedding_vector = word2vec_model.wv[word]
            embedding_matrix[i] = embedding_vector
        except KeyError:
            pass  # Words not found in the embedding index will be all zeros
    return embedding_matrix

eng_embedding_matrix = create_embedding_matrix(eng_model, english_tokenizer, english_vocab_size)
tam_embedding_matrix = create_embedding_matrix(tam_model, tamil_tokenizer, tamil_vocab_size)
```

# Create SequencetoSequence model
### Workflow for building an encoder for Seq2Seq model
1.Input Layer: Takes a sequence of tokenized words (e.g., an English sentence

2.Embedding Layer: Converts each word into a dense vector using pre-trained embeddings (from eng_embedding_matrix).

3.LSTM Layer: Processes the sequence of vectors to generate context (hidden and cell states) to be passed to the decoder.

This is the encoder part of a sequence-to-sequence architecture.

```python
encoder_inputs = Input(shape=(20,))
encoder_embedding = Embedding(english_vocab_size, 100, weights=[eng_embedding_matrix], trainable=False)(encoder_inputs)
encoder_lstm, encoder_state_h, encoder_state_c = LSTM(100, return_state=True)(encoder_embedding)
```

```python
#function to create seq2seq model
def create_seq2seq_model(input_vocab_size, output_vocab_size, input_seq_length, output_seq_length, hidden_units, eng_embedding_matrix, tam_embedding_matrix):
    # Encoder
    encoder_inputs = Input(shape=(input_seq_length,))
    encoder_embedding = Embedding(input_vocab_size, hidden_units, weights=[eng_embedding_matrix], trainable=False)(encoder_inputs)
    encoder_lstm, encoder_state_h, encoder_state_c = LSTM(hidden_units, return_state=True)(encoder_embedding)

    # Decoder
    decoder_inputs = Input(shape=(output_seq_length,))
    decoder_embedding = Embedding(output_vocab_size, hidden_units, weights=[tam_embedding_matrix], trainable=False)(decoder_inputs)
    decoder_lstm = LSTM(hidden_units, return_sequences=True, return_state=True)
    decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=[encoder_state_h, encoder_state_c])
    decoder_dense = Dense(output_vocab_size, activation='softmax')
    decoder_outputs = decoder_dense(decoder_outputs)

    model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
    return model
```

```python
target_sequences = tf.keras.utils.to_categorical(output_sequences, num_classes=tamil_vocab_size)
```

```python
# Preprocessing the input
input_sentence = "<sosThe man came to give his paper yesterday evening and i was busy on that time<eos>"

# Convert the input sentence to sequence
input_sequence = english_tokenizer.texts_to_sequences([input_sentence])

# Pad the statement to the maximum input sequence length
input_sequence_pad = pad_sequences(input_sequence, maxlen=max_input_seq_length, padding='post')

# Generate predictions
predictions = model.predict([input_sequence_pad, np.zeros((1, max_output_seq_length))])

# Convert predictions to tokens
predicted_tokens = np.argmax(predictions, axis=-1)[0]

# Create index to word mapping for Tamil vocabulary
tamil_index_word = {i: w for w, i in tamil_tokenizer.word_index.items()}


# Convert tokens to text
decoded_sentence = []
for token in predicted_tokens:
    if token == 0:  # Assuming 0 is the padding token
        continue
    word = tamil_index_word.get(token)
    if word == '<eos>':
        break
    if word is not None:
        decoded_sentence.append(word)
    else:
        decoded_sentence.append('<unk>')

# Join the words to form the decoded statement
decoded_statement = ' '.join(decoded_sentence)

# Print the decoded statement
print(decoded_statement)
```

## How the predicted tokens are converted into its corresponding words?
### 1. An empty list is created
### 2. Loop iterates over each token in predicted_tokens.
### 3. if the token is 0, which is assumed to be the padding token. 
### 4. Corresponding word for the token index is retrieved
### 5. If the decoded word is '<eos>' (end of sentence), the loop breaks, meaning it stops decoding further.
### 6. Checks if a valid word is found for the token,if the word is found, it appends the word to decoded_sentence otherwise it appends '<unk>' to the sentence.

```python
decoded_sentence = []
for token in predicted_tokens:
    if token == 0:  # Assuming 0 is the padding token
        continue
    word = tamil_index_word.get(token)
    if word == '<eos>':
        break
    if word is not None:
        decoded_sentence.append(word)
    else:
        decoded_sentence.append('<unk>')
```
